```python
import sys

import mediapy
import yaml
from tqdm import tqdm

sys.path.append("../")
```

```python
from lcsim.engine import Engine
```

specify config

```python
config = dict(yaml.safe_load(open("../lcsim/config/example.yml", "r")))
config["input"]["data_dir"] = "data/waymo/1a1b6e6bacdd8880"
```

init simulation engine

```python
engine = Engine(config)
```

run simulation and show result

```python
imgs = []
render_interval = 1
engine.reset()
for i in tqdm(range(90)):
    engine.step()
    if i % render_interval == 0:
        img = engine.render()
        if not img is None:
            imgs.append(img)
        else:
            break

mediapy.show_video(imgs, fps=24)
```
